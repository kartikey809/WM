iii. Apply Elias Gamma Encoding for term “Airtel”

```python
from math import log
log2 = lambda id: log(id, 2)

def UnaryConv(n):  #unary for N
	return (n-1)*'0'+'1'

def BinaryConv(b, l = 1): #inary conversion
	s = '{0:0%db}'% l
	return s.format(b)
	
def Elias_Gamma(id): 
	if(id == 0):
		return '0'
	n = 1 + int(log2(id))  # n >  log2(num) where n is nearest 2^n < num
	b = id - 2**(int(log2(id)))   # X- 2^N
	l = int(log2(id))
	return UnaryConv(n) + BinaryConv(b, l)
    
Airtel = [12, 17, 25, 148, 156, 159, 172]
print("Elias Delta Encoding for Airtel doc ids = ", end=" ")
for n in Airtel:
    print(Elias_Gamma(n), end=" ")
```

iv. Apply Elias Delta Decoding for “000010000”

```python
import math
  
  
def EliasDeltaDecoding(x):
    x = list(x)
    L = 0
    while True:
        if not x[L] == '0':
            break
        L = L + 1 #converting unary to the number
          
    # Reading L more bits and dropping ALL    
    x = x[2*L+1:]  
      
    # adding 1 at front
    x.reverse()
    x.insert(0, '1')  
    
      
    # Converting binary to integer
    n=0
    for i in range(len(x)):  
        if x[i] == '1':
            n = n+math.pow(2, i)
    return int(n)
  
  
x = '000010000'
print("Decoded value is ",EliasDeltaDecoding(x))
```

v. Apply Elias Delta Encoding for term “Venus”

```python
from math import log
from math import floor

def Offset(x): #removing 1 in MSB
	binary = "{0:b}".format(int(x))
	offset = binary[1:]
	return offset

def EliasGammaEncode(k):
	if (k == 0):
		return '0'
	N = 1 + floor(log(k, 2))
	Unary = (N-1)*'0'+'1'
	return Unary + Offset(k) #concatenating selector and offset

def EliasDeltaEncode(n):
	Gamma = EliasGammaEncode(1 + floor(log(n, 2)))
	offset = Offset(n)
	return Gamma+offset 

print("Elias Delta Encoding for Venus Posting List ids = ", end=" ")
Venus = [23, 45, 78, 122, 145]
for id in Venus:
	print(EliasDeltaEncode(id), end=" ")
```

vii. Apply Variable Byte Encoding for “Samsung”. (Use doc ids gap)

```python
def toBinary(number):
    bin_num = bin(number)
    bin_num = bin_num[2:]
    return bin_num

def variableByteEncoding(number):
    bin_code = toBinary(number)
    vb_enc = ""

    while len(bin_code) > 0:
        if len(bin_code) > 7:
            term = bin_code[-7:]  #getting last 7 bits
            bin_code = bin_code[:-7] #remaining bits
        else:
            term = bin_code
            bin_code = ""
            term = ("0" * (7 - len(term))) + term #prepending 0s  

        if len(vb_enc) == 0:
            vb_enc = term + "0"
        else:
            vb_enc = term + "1" + vb_enc
    return vb_enc

print("Variable Byte Encoding for Samsung = ", end=" ")
Samsung = [2, 12, 34544, 34574, 35569]
for i in range(len(Samsung)-1):
    if i == 0:
        print(variableByteEncoding(Samsung[i]), end=" ")
    else:
        x = int(variableByteEncoding(Samsung[i+1]))
        y = int(variableByteEncoding(Samsung[i]))
        print(x - y, end=" ")
```
