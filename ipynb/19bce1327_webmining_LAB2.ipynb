##a. Breadth-First-Search

```python
import requests
from bs4 import BeautifulSoup #for parsing HTML and XML content in webpages
from collections import deque
import re
url = input("enter the source url :")
visitedlinks = set(url)
Queue = deque([[url, "", 0]])
max_depth = 10
search_word = input("enter the search word :")
while Queue:
    base, path, depth = Queue.popleft() #while queue is not empty popping elements

    if depth < max_depth:
        try:
            page = BeautifulSoup(requests.get(base + path).text, "html.parser")

            for link in page.find_all("a"): #find all anchor tags links in html page
                href = link.get("href") #getting href link

                if href not in visitedlinks:
                    visitedlinks.add(href) #adding link to visited set
                    print("  " * depth + f"depth {depth} -- {href}")
                    if(re.search(search_word, href)):   #searching the word in href link
                        if href.startswith('https'):
                            Queue.append([href, "", depth + 1]) #appending in queue
                        else:
                            Queue.append([base, href, depth + 1])
					
        except:
            pass
```

##Depth-First Search

```python
def search_dfs(base, path, visited, searchword,max_depth=3, depth=0):
    if depth < max_depth:
        try:
            soup = BeautifulSoup(requests.get(base + path).text, "html.parser")
            for link in soup.find_all("a"): #finding all anchor tag links
                href = link.get("href") 
                if href not in visited:
                    visited.add(href) #adding to visited pages
                    print(f"at depth {depth}: {href}")
                    if(re.search(searchword, href)):  #searching for keyword in href links
                        if href.startswith("http"):
                            search_dfs(href, "", visited, max_depth-1, depth + 1) #calling recursively
                        else:
                            search_dfs(base, href, visited, max_depth-1, depth + 1) #calling recursively
        except:
            pass
link=input("Enter the link to search from: ")
searchword = input("enter search word :")
search_dfs(link, "", set([link]),searchword)  
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
```

```python

from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
driver.get('https://chennai.vit.ac.in/about/contactus/') 
chennai = driver.find_elements_by_xpath('/html/body/div[1]/div/div[2]/div/div/div/article/div[2]/div[1]') #xpath found using chrome developer tools
chennai_list = [] 
for p in range (len(chennai)): 
    chennai_list.append(chennai[p].text) #storing text in html in list obtained from web driver
    print(chennai_list[p])
```
